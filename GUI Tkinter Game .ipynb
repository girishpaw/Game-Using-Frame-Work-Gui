```python
from tkinter import *
import random
root = Tk()
root.geometry("500x600")
root.title("Rock Paper Scissor Game")
computer_value = {"0":"Rock","1":"Paper","2":"Scissor"}

def rock():
    cv = computer_value[str(random.randint(0,2))]
    if cv == "Rock":
        match_result = "Match Draw"
    elif cv=="Scissor":
        match_result = "Player1 Win"
    else:
        match_result = "player2 Win"
    l4.config(text = match_result)
    l1.config(text = "Rock")
    l3.config(text = cv)
    button_disable()

def paper():
    cv = computer_value[str(random.randint(0, 2))]
    if cv == "Paper":
        match_result = "Match Draw"
    elif cv=="Scissor":
        match_result = "player2 Win"
    else:
        match_result = "Player1 Win"
    l4.config(text = match_result)
    l1.config(text = "Paper")
    l3.config(text = cv)
    button_disable()

def scissor():
    cv = computer_value[str(random.randint(0,2))]
    if cv == "Rock":
        match_result = "player2 Win"
    elif cv == "Scissor":
        match_result = "Match Draw"
    else:
        match_result = "Player1 Win"
    l4.config(text = match_result)
    l1.config(text = "Scissor")
    l3.config(text = cv)
    button_disable()

def reset_game():
    b1["state"] = "active"
    b2["state"] = "active"
    b3["state"] = "active"
    l1.config(text = "Player")
    l3.config(text = "Computer")
    l4.config(text = "")

def button_disable():
    b1["state"] = "active"
    b2["state"] = "active"
    b3["state"] = "active"

Label(root,text="Rock Paper Scissor",font="Arial_Black",bg="green").pack(pady=24)

frame=Frame(root)
frame.pack()

l1=Label(frame,text="Player",font = "normal 10 bold")
l1.pack(side=LEFT)

l2 = Label(frame,text = "VS",font = "normal 10 bold")
l2.pack(side = LEFT)

l3 = Label(frame, text = "Computer", font = "normal 10 bold")
l3.pack(side = LEFT)

l4 = Label(root,text = "",font ="normal 20 bold",bg = "white",width = 15 ,borderwidth = 2,relief = SUNKEN)
l4.pack(pady = 20)

frame1=Frame(root)
frame1.pack()

b1 = Button(frame1, text = "Rock",font = 10, width = 7,command = rock)
b1.pack(side = LEFT, padx = 10)

b2 = Button(frame1, text = "Paper ",font = 10, width = 7,command = paper)
b2.pack(side = LEFT,padx = 10)

b3 = Button(frame1, text = "Scissor",font = 10, width = 7,command = scissor)
b3.pack(padx = 10)

Button(root,text="Reset Game",font = 20, fg="Green",bg="Red",command=reset_game).pack(pady=15)

root.mainloop()
```

```python
import tkinter as tk  
  
root = tk.Tk()  
#root.resizable(False, False)
root.geometry("500x600")
root.title("Tic Tac Toe")  
  
tk.Label(root, text="Tic Tac Toe", font=('Ariel', 25)).pack()
current_chr = "X"

status_label = tk.Label(root, text="X's turn", font=('Ariel', 15), bg='green', fg='snow')

play_area = tk.Frame(root, width=300, height=300, bg='white')  
X_points = []
O_points = []
class XOPoint:  
    def __init__(self, x, y):
        self.x = x  
        self.y = y  
        self.value = None  
        self.button = tk.Button(play_area, text="", width=10, height=5, command=self.set)
        self.button.grid(row=x, column=y)

    def set(self):
        global current_chr
        if not self.value:
            self.button.configure(text=current_chr, bg='snow', fg='black')
            self.value = current_chr 
            if current_chr == "X":
                X_points.append(self)
                current_chr = "O"
            elif current_chr == "O":
                O_points.append(self)
                current_chr = "X"
        check_win()
  
    def reset(self):  
        self.button.configure(text="", bg='white')  
        self.value = None
for x in range(1, 4):
    for y in range(1, 4):
        XOPoint(x, y)
if len(X_points) + len(O_points) == 9:
        print("Draw!")
class WinningPossibility:
    def __init__(self, x1, y1, x2, y2, x3, y3):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2
        self.x3 = x3
        self.y3 = y3
    def check(self, for_chr):
        p1_satisfied = False
        p2_satisfied = False
        p3_satisfied = False
        if for_chr == 'X':
            for point in X_points:
                if point.x == self.x1 and point.y == self.y1:
                    p1_satisfied = True
                elif point.x == self.x2 and point.y == self.y2:
                    p2_satisfied = True
                elif point.x == self.x3 and point.y == self.y3:
                    p3_satisfied = True
        elif for_chr == 'O':
            for point in O_points:
                if point.x == self.x1 and point.y == self.y1:
                    p1_satisfied = True
                elif point.x == self.x2 and point.y == self.y2:
                    p2_satisfied = True
                elif point.x == self.x3 and point.y == self.y3:
                    p3_satisfied = True
        return all([p1_satisfied, p2_satisfied, p3_satisfied])
winning_possibilities = [
    WinningPossibility(1, 1, 1, 2, 1, 3),
    WinningPossibility(2, 1, 2, 2, 2, 3),
    WinningPossibility(3, 1, 3, 2, 3, 3),
    WinningPossibility(1, 1, 2, 1, 3, 1),
    WinningPossibility(1, 2, 2, 2, 3, 2),
    WinningPossibility(1, 3, 2, 3, 3, 3),
    WinningPossibility(1, 1, 2, 2, 3, 3),
    WinningPossibility(3, 1, 2, 2, 1, 3)
]
def check_win():
    for possibility in winning_possibilities:
        if possibility.check('X'):
            print("X won!")
            return
        elif possibility.check('O'):
            print("O won!")
            return
play_area.pack(pady=10, padx=10)  

root.mainloop()
```

```python
import tkinter
import random

colours = ['Red','Blue','Green','Pink','Black','Yellow','Orange','White','Purple','Brown']
score = 0

timeleft = 30

def startGame(event):
     
    if timeleft == 30:
        countdown()
    nextColour()

def nextColour():
    global score
    global timeleft
    if timeleft > 0:
        e.focus_set()
        if e.get().lower() == colours[1].lower():
            score += 1
        e.delete(0, tkinter.END)
        random.shuffle(colours)
        label.config(fg = str(colours[1]), text = str(colours[0]))
        scoreLabel.config(text = "Score: " + str(score))

def countdown():
    global timeleft
    if timeleft > 0:
        timeleft -= 1
        timeLabel.config(text = "Time left: "+ str(timeleft))
        timeLabel.after(1000, countdown)

root = tkinter.Tk()

root.title("COLORGAME")

root.geometry("375x200")
 
instructions = tkinter.Label(root, text = "Type in the colour""of the words, and not the word text!",font = ('Helvetica', 12))
instructions.pack()

scoreLabel = tkinter.Label(root, text = "Press enter to start",font = ('Helvetica', 12))
scoreLabel.pack()

timeLabel = tkinter.Label(root, text = "Time left: " +str(timeleft), font = ('Helvetica', 12))
               
timeLabel.pack()

label = tkinter.Label(root, font = ('Helvetica', 60))
label.pack()
 
e = tkinter.Entry(root)

root.bind('<Return>', startGame)
e.pack()

e.focus_set()

root.mainloop()
```
